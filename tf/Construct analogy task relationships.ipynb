```python
import music21
```

```python
def cycle(a, b, delta=7):

    SEED_A = music21.chord.Chord(a)
    SEED_B = music21.chord.Chord(b)

    cycle = []

    for i in range(12):
        ent_1 = SEED_A.transpose(i)
        ent_2 = SEED_B.transpose(i + delta)
        ent_3 = SEED_A.transpose(i + delta)
        ent_4 = SEED_B.transpose(i + (delta*2))
        row = [ent_1, ent_2, ent_3, ent_4]

        def stringify(c):
            return "".join(str(c.orderedPitchClasses)).replace(' ', '')
        cycle.append(map(stringify, row))

    return cycle
```

```python
relationships = []

# Dominant 7th to 5th-related major (V7-I)
relationships.extend(cycle('C E G B-', 'C E G',delta=7))

# Dominant 7th to 5th-related minor (V7-i)
relationships.extend(cycle('C E G B-', 'C E- G',delta=7))

# Major to 5th-related major (V-I)
relationships.extend(cycle('C E G', 'C E G',delta=7))

# Major to 5th-related minor (V-i)
relationships.extend(cycle('C E G', 'C E- G',delta=7))
```

```python
with open('pc_eval.txt', 'w') as f:
    f.writelines(map(lambda x: " ".join(x) + "\n", relationships), )
```
